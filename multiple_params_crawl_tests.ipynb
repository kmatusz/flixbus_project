```python
from itertools import product
search_params = {
    "departureCity": "7568",
    "arrivalCity": "1915",
    "rideDate": "22.11.2019",
    "adult": "1",
    "_locale": "pl",
}
```

```python
rideDate = pd.date_range(pd.datetime.today(), periods=50).strftime('%d.%m.%Y').tolist()[1:]
```

```python
cities = {
    #"Warszawa": 7568,
    "Krakow": 1915,
    "Gdansk": 1815,
    "Lodz": 7588,
    "Wroclaw": 1765
}
```

```python
from itertools import product
params_list = []

for i in product(cities.values(), rideDate):

    params_list.append({
    "departureCity": "7568",
    "arrivalCity": i[0],
    "rideDate": i[1],
    "adult": "1",
    "_locale": "pl",
})
    
params_df = pd.DataFrame(params_list)
```

```python
params_df
```

```python
%time
results_df_list = []

params_list_from_df = params_df.to_dict("index")

for key, value in params_list_from_df.items():
    print(key)
    crawler = Crawler(value)
    crawler.get_all()
    
    results = crawler.results
    temp_df = pd.DataFrame(results)
    temp_df["params_idx"] = key
    results_df_list.append(temp_df)
    
```

```python
all_results = pd.concat(results_df_list)
```
